```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Finance_data.csv')
```

BASIC GRAPH TO SHOW GENDER DISTRIBUTION THAT TAKES THE SURVEY

```python
gender_counts = df['gender'].value_counts()

# Plot a bar chart
gender_counts.plot(kind='bar', color=['blue', 'pink'], edgecolor='black')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')

# Show the plot
plt.show()
```

BASIC STACKED BAR CHART TO SHOW GENDER DISTRIBUTION WHETHER OR NOT THEY INVEST IN AVENUES

```python
gender_investment_counts = df.groupby(['gender', 'Investment_Avenues']).size().unstack()

# Plotting the stacked bar chart
ax = gender_investment_counts.plot(kind='bar', stacked=True, color=['skyblue', 'yellow'])

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Relationship between Gender and Investment Avenue')

# Adding legend
plt.legend(title='Investment Avenue', loc='upper right')

# Display the chart
plt.show()
```

BASIC TO SHOW HOW FREQUENT THEY(THE ONE THAT ANSWERS THE SURVEY) MONITOR THEIR INVESMENT

```python
sns.countplot(x=df['Duration'], hue=df['Invest_Monitor'], palette=['blue', 'orange','green'], edgecolor='black')
plt.legend(loc='upper right')
plt.show()
```

BASIC STACKED BAR CHART TO SHOW INVESMENT OBJECTIVES BY GENDER

```python
# Create a DataFrame for the stacked bar chart
df_stacked = df.groupby(['gender', 'Objective']).size().unstack().fillna(0)

# Stacked bar chart with numbers on bars
ax = df_stacked.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Adding labels, title, and legend
ax.set_ylabel('Count')
ax.set_xlabel('Gender')
ax.set_title('Stacked Bar Chart of Investment Objectives by Gender')

# Adding numbers on each section of the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', color='black', size=10)

plt.show()
```

BASIC STACKED BAR CHART TO SHOW WHERE THEY GET INFORMATION REGARDING INVESMENT BY GENDER

```python
# Create a DataFrame for the stacked bar chart
df_stacked = df.groupby(['gender', 'Source']).size().unstack().fillna(0)

# Stacked bar chart with numbers on bars
ax = df_stacked.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='inferno')

# Adding labels, title, and legend
ax.set_ylabel('Count')
ax.set_xlabel('Gender')
ax.set_title('Stacked Bar Chart of where they get information regarding invesment by Gender')

# Adding numbers on each section of the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', color='black', size=10)

plt.show()
```
